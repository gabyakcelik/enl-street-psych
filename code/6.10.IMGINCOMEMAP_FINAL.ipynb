```python
from cartopy.io import shapereader
import os
import shapefile
from json import dumps
import numpy as np
from IPython.display import IFrame
import pandas as pd
import folium
import geopandas as gpd 
import numpy as np
from folium.plugins import MarkerCluster
```

```python
tract_map = gpd.read_file('/content/geo_export_cc46e642-4ec3-4e7f-8a91-95e2bded87b3.shp')
tract_map.to_file('comm_map.gejson', driver='GeoJSON')
```

```python
tract_map['name10'] = tract_map['name10'].infer_objects()
tract_map.dtypes
tract_map['name10']  = tract_map.name10.astype(float)
tract_map.dtypes
```

```python
sidegeo = pd.read_csv("https://raw.githubusercontent.com/gabyakcelik/street_psych/master/Geocode/side_geoids_updated.csv")
sidegeo = sidegeo.drop(['Unnamed: 6'], axis=1)
census = pd.read_csv("https://raw.githubusercontent.com/gabyakcelik/street_psych/master/Geocode/updated_geoid_census2019.csv")
census = census.rename(columns={'GEOID_text' : 'block_group_code'})
sw_cen = pd.merge(sidegeo, census, on=['block_group_code'])
```

```python
cen = pd.read_csv("https://raw.githubusercontent.com/gabyakcelik/street_psych/master/ReExtracted_Data_6.7.2021/img_income_comm_geocord_sw.csv")
cen = cen.rename(columns={'Community Area Name' : 'community', 'Fixed Tract Match' : 'tractce10'})
cen['name10']
```

## Census Tracts Map

```python
m1 = pd.merge(tract_map, cen, on=['name10'], how='left')
```

```python
m1 = m1[['name10','MedianHHIncome', 'geometry' ]]
```

```python
m1
```

```python
m1 = gpd.GeoDataFrame(m1)
```

```python
hhi = sw_cen[['a_image_name', 'block_group_code', 'lat', 'lng','MedianHHIncome']]
hhi = hhi.loc[hhi["MedianHHIncome"] != -666666666]
hhi
```

```python
hhi['a_image_name'][378]
```

```python
hhi.iloc[378]
```

```python
hhi['a_image_name'][94]
```

```python
imglocations = hhi[['lat', 'lng']]
imglocations
locationlist = imglocations.values.tolist()
len(locationlist)
locationlist[94]
```

```python
hhi.iloc[94]
```

```python
map1 = folium.Map(location=[41.95781406, -87.67870071]
,tiles = 'cartodbpositron', zoom_start=20, width=900, height=900)
```

```python
folium.Choropleth( 
    geo_data=m1,
    data=m1,
    columns=['name10','MedianHHIncome'],
    key_on='feature.properties.name10',
    fill_color='BuPu',
    fill_opacity=.5,
    line_opacity=1,
    legend_name="Median Household Income",
    smooth_factor=2.0,
    Highlight= True,
    line_color = "black",
    name = "Median Household Income",
    show=False,
    overlay=True,
    nan_fill_color = "black"
).add_to(map1)

map1
```

```python
marker_cluster = MarkerCluster().add_to(map1)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=sw_cen['a_image_name'][point], 
                  icon=folium.Icon(color='darkblue', icon_color='black', icon='fa-camera', angle=0, prefix='fa')).add_to(marker_cluster)


map1.save('map1.html')
```
